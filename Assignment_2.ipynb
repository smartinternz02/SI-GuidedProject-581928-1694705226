```python
#Task-1:Download the data set
#Task-2:Load the dataset

import pandas as pd
df=pd.read_csv('/content/House Price India.csv')
df
```

```python
#Task-3:Univariate analysis

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
sns.distplot(df.Lattitude)
```

```python
#line plot
df.Date.plot()
plt.legend(['date'])
```

```python
#pie chart
df1=df.head(10)
plt.pie(df1.lot_area_renov,autopct='%1.1f%%')
plt.title('lot_area_renov')
plt.show()
```

```python
#bar graph
plt.bar(df1.index,df1.Price)
```

```python
#scatter graph
plt.scatter(df1.index,df1.living_area_renov)
```

```python
#Bivariate graph (joint plot)
df2=df.head(100)
sns.jointplot(x='Lattitude',y='Longitude',data=df2)
```

```python
#multi variate graph(heat map)
sns.heatmap(df.corr())
```
